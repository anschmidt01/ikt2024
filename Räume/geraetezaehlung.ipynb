```python
import pandas as pd
import matplotlib.pyplot as plt
from data_analysis import load_data, plot_data

# Pfad zur CSV-Datei
file_path = '2024-05-18_Export_01_DRA.csv'  # Ersetze dies durch den tatschlichen Pfad zur Datei

# CSV-Datei laden
data = load_data(file_path)

# Rume, die analysiert werden sollen
rooms = ['Raum625', 'Raum639', 'Raum640']

for room in rooms:
    room_data = data[data['Room'] == room]
    plot_title = f'Gertezhlung ber die Zeit in {room}'
    plot_file_path = f'geraetezaehlung_ueber_zeit_{room}.png'
    plot_data(room_data, plot_title, plot_file_path)
```
